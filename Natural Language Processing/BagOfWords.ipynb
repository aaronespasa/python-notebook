```
from sklearn.feature_extraction.text import CountVectorizer
# More info about CountVectorizer:
# https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html
import pandas as pd
```

```
documents = ['Hello, how are you!',
             'Win money, win from home.',
             'Call me now.',
             'Hello, Call hello you tomorrow?']
```

```
# Convert documents into a CountVector
count_vector = CountVectorizer(documents)

# Information about the count_vector object
print(count_vector)
```

```
# Fit method allows us to learn the vocabulary
count_vector = count_vector.fit(documents)

# Transform method convert documents into document-term matrix
# Finally, we convert the object into an array with .toarray()
doc_array = count_vector.transform(documents).toarray()
print(doc_array)
```

```
# Convert the array into a more legible format
df = pd.DataFrame(doc_array, columns=count_vector.get_feature_names())
print(df)
```
