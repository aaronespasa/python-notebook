## Reading common file formats

### CSV

A CSV file is a table of values separated by commas. Hence the name: "*comma-seperated values*"

```python
import pandas as pd

wine_reviews = pd.read_csv("./csv/winemag-data-130k-v2.csv")
```

<code>shape</code> attribute is used to check how large the retulting DataFrame is

```python
wine_reviews.shape
```

<code>head</code> command grabs the first five rows

```python
wine_reviews.head()
```

To make pandas use that column for the index (instead of creating a new one from scratch), we may specify and use an <code>index_col</code>.

```python
wine_reviews = pd.read_csv("csv/winemag-data-130k-v2.csv", index_col = 0)
wine_reviews.head()
```

### Excel (XLS or XLST):
The spreadsheet organizes itself as a sequence of named sheets

```python
wic = pd.read_excel("xls_files_all/WICAgencies2013ytd.xls",
                   sheet_name='Total Women')
wic.head()
```

### SQL files:

SQL databases are where most of the data on the web ultimately gets stored

```python
import sqlite3
conn = sqlite3.connect("sqlite/FPA_FOD_20170508.sqlite")
```

Internally, SQL databases all operate very differently. Externally, however, they all provide the same API, the "Structured Query Language" 

```python
fires = pd.read_sql_query("SELECT * FROM fires", conn)
```

```python
fires.head()
```

## Writing common file formats

The opposite of read_csv, which reads our data, is <code>to_csv</code>, which writes it.

```python
wine_reviews.head().to_csv("csv/wine_reviews.csv")
```

To write an Excel file back you need <code>to_excel</code> and the <code>sheet_name</code> again:

```python
wic.to_excel("xls_files_all/wic.xlsx", sheet_name="Total Women")
```

To output to a SQL database, supply the name of the table in the database we want to throw the data into, and a connector:

```python
conn = sqlite3.connect("fires.sqlite")
fires.head(10).to_sql("sqlite/fires", conn)
```
